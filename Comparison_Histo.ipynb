```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ng_epta_snr=np.load("/Users/physics/nanogw/SNR_mar_data/ng_epta_snr.npy")
ng_snr=np.load("/Users/physics/nanogw/SNR_mar_data/ng_snr.npy")
ng_mar_snr=np.load("/Users/physics/nanogw/SNR_mar_data/ng_mar_snr.npy")
epta_snr=np.load("/Users/physics/nanogw/SNR_mar_data/epta_snr.npy")
epta_mar_snr=np.load("/Users/physics/nanogw/SNR_mar_data/epta_mar_snr.npy")
```

```python
plt.hist(ng_epta_snr, histtype='step', lw=2, label='NANOGrav and EPTA', color='C0',
         bins=20, density=True)
plt.hist(ng_snr, histtype='step', lw=2, label='NANOGrav', color='C1',
         bins=20, density=True)
plt.hist(ng_mar_snr, histtype='step', lw=2, label='NANOGrav Marginalized', color='C2',
         bins=20, density=True)
plt.hist(epta_snr, histtype='step', lw=2, label='EPTA', color='C3',
         bins=20, density=True)
plt.hist(epta_mar_snr, histtype='step', lw=2, label='EPTA Marginalized', color='C4',
         bins=20, density=True)

plt.gca().axvline(x=np.mean(ng_epta_snr), ls='--', color='C0')
plt.gca().axvline(x=np.mean(ng_snr), ls='--', color='C1')
plt.gca().axvline(x=np.mean(ng_mar_snr), ls='--', color='C2')
plt.gca().axvline(x=np.mean(epta_snr), ls='--', color='C3')
plt.gca().axvline(x=np.mean(epta_mar_snr), ls='--', color='C4')


plt.legend(loc="upper right")
plt.xlabel("S/N")
plt.ylabel("PDF")
plt.savefig("EPTA_NANOGrav.pdf")
```
